```python
def get_feature_map(index='JOND920101'):
    """ To get the feature mapping object. """
    
    # Create a Feature object
    aaindex_map = get_aaindex_file(index)
    aaindex_map.mapping['-'] = 0.0
    feat_map = Feature(aaindex_map)
    
    return feat_map
```
